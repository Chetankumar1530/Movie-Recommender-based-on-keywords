```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
from IPython.display import clear_output, Image
```

```python
data = pd.read_csv("Data With NLP.csv")
```

```python
data["img_url"] = 0
```

```python
data["IMDB ID"][0]
```

```python
error = 0
for i in range(0,len(data)):
    try:
        a= data["IMDB ID"][i]
        url = f"https://www.imdb.com/title/{a}/"
        d = requests.get(url)
        d = bs(d.content)
        img = d.find(class_="ipc-lockup-overlay ipc-focusable")["href"]
        img = "https://www.imdb.com"+img
        d = requests.get(img)
        d = bs(d.content)
        data["img_url"][i]= (d.find(class_="hXPlvk")["src"])
        

        clear_output(wait=True)
        display(Image(data["img_url"][i],width=250))
    except:
        data["img_url"][i]= "sorry"
        error += 1
    print(round((i/len(data))*100,2),"% Completed")
    print(error ,"Error Found")
```

```python
data.to_csv("Nlp with images.csv")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
import ast
import warnings
warnings.filterwarnings("ignore")
vec = TfidfVectorizer()
data = pd.read_csv("Nlp with images.csv")
data.fillna("sorry",inplace=True)
vector = vec.fit_transform(data["ggg"])
indexed = pd.Series(data = data.index,index=data.lower)

def search(n, top = 10,rate = 3):
    if n.lower() in indexed:
        rec = pd.DataFrame(list(linear_kernel(vector[indexed[n]],vector)))
        rec = rec.transpose()
        rec.columns = ["name"]
        rec = rec.sort_values(by="name" ,ascending=False)
        x = pd.DataFrame()
        for i in range(0,min(top,800)):
            x = x.append(data.loc[rec.index[i]:rec.index[i]])
        x = x[x["Ratinng"] > rate]
        x = x.sort_values(by="Ratinng",ascending=False)
        
#         display(x)
    else:
        return ("Not Found")
    return list(x["Movie Title"]), list(x.ggg), list(x.Ratinng),list(x["Plot"]),list(x["Positive reviews"]),list(x["Negative reviews"]),list(x["Nuteral  reviews"]),list(x["img"])


def search_by_keyword(n, top = 10,rate = 3):
    data.Plot.loc[810] = n
    vector = vec.fit_transform(data["Plot"])
    indexed = pd.Series(data = data.index,index=data.title)

    rec = pd.DataFrame(list(linear_kernel(vector[indexed[810]],vector)))
    rec = rec.transpose()
    rec.columns = ["name"]
    rec = rec.sort_values(by="name" ,ascending=False)
    x = pd.DataFrame()
    for i in range(1,min(top,800)):
        x = x.append(data.loc[rec.index[i]:rec.index[i]])
    x = x[x["Ratinng"] > rate]
    return list(x["Movie Title"]), list(x.ggg), list(x.Ratinng),list(x["Plot"]),list(x["Positive reviews"]),list(x["Negative reviews"]),list(x["Nuteral  reviews"]),list(x["img"])


```
